```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns',None)
import warnings
warnings.filterwarnings("ignore")
```

# 特征筛选

```python
origin=pd.read_csv('../data(1).csv')
origin
```

## 特征筛选

```python
#需要的列
col=['loan_amnt','term','int_rate','installment','grade','sub_grade','emp_title','emp_length','home_ownership','annual_inc',
    'issue_d','loan_status','purpose','addr_state','dti','delinq_2yrs','fico_range_low','fico_range_high',
    'revol_bal','acc_now_delinq','acc_open_past_24mths','delinq_amnt','verification_status',
    'pct_tl_nvr_dlq','pub_rec_bankruptcies','hardship_flag','disbursement_method']

#提取出有用的列
data=origin[col]
```

## 类别型feature转换为数字

```python
#term
dic={' 36 months':36,' 60 months':60}
data['term']=data['term'].map(dic)
#emp_length
dic={'1 years':1,'2 years':2,'3 years':3,'4 years':4,'5 years':5,'6 years':6,'7 years':7,'8 years':8,
     '9 years':9,'< 1 years':0,'10+ years':10
    }
data['emp_length']=data['emp_length'].map(dic)
#disbursement_method
dic={'Cash':0,'DirectPay':1}
data['disbursement_method']=data['disbursement_method'].map(dic)
#hardship_flag
dic={'N':0,'Y':1}
data['hardship_flag']=data['hardship_flag'].map(dic)
#grade
dic={'A':1,'B':2,'C':3,'D':4,'E':5,'F':6,'G':7}
data['grade']=data['grade'].map(dic)
#sub_grade
dic={'A1':1,'A2':2,'A3':3,'A4':4,'A5':5,'B1':6,'B2':7,'B3':8,'B4':9,'B5':10,'C1':11,'C2':12,'C3':13,'C4':14,'C5':15,
    'D1':16,'D2':17,'D3':18,'D4':19,'D5':20,'E1':21,'E2':22,'E3':23,'E4':24,'E5':25,
    'F1':26,'F2':27,'F3':28,'F4':29,'F5':30,'G1':31,'G2':32,'G3':33,'G4':34,'G5':35}
data['sub_grade']=data['sub_grade'].map(dic)
```

# 数值型的feature，做归一化


```python
number_col=['loan_amnt','term','int_rate','installment','emp_length','annual_inc','dti','delinq_2yrs','fico_range_low','fico_range_high',
         'revol_bal','acc_now_delinq','acc_open_past_24mths','delinq_amnt','pct_tl_nvr_dlq','pub_rec_bankruptcies','hardship_flag',
         'disbursement_method']
#loan_amnt
df=data[number_col]
df = (df-df.min())/(df.max()-df.min())
df
```

# 类别型的feature，做one-hot编码，然后乘以映射矩阵W变为dense vector作为特征

```python
from sklearn import preprocessing
enc = preprocessing.OneHotEncoder(handle_unknown='ignore')
text_col=['home_ownership','loan_status','purpose','addr_state','verification_status']#文本型的列
```

```python
#one-hot编码
X=pd.get_dummies(data[text_col])
```

```python
#final dataset--用于训练的dataset
final=pd.concat([df,X],axis=1)
final=final.fillna(0)
final=final.to_numpy()
```

```python
####pub_rec
result=origin['pub_rec']
dic={0:0,1:1,2:1,3:1,4:1,5:1,6:1,7:1,8:1,10:1}#转化成2类
result=result.map(dic)
```

## 训练模型

```python
#####开始训练
#####knn
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
X_train,X_test,y_train,y_test=train_test_split(final,result,test_size=0.2)
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)#training的过程

#评分
print('knn score: ',knn.score(X_test,y_test))
```

```python
####svm
from sklearn.svm import SVC
X_train,X_test,y_train,y_test=train_test_split(final,result,test_size=0.2)
clf=SVC()
clf.fit(X_train,y_train)
print('svm score: ',clf.score(X_test,y_test))
```

```python
####神经网络
from sklearn.neural_network import MLPClassifier
X_train,X_test,y_train,y_test=train_test_split(final,result,test_size=0.2)
clf = MLPClassifier(random_state=1, max_iter=150).fit(X_train, y_train)
print('mlp score: ',clf.score(X_test,y_test))
```

```python
####逻辑回归
from sklearn.linear_model import LogisticRegression
X_train,X_test,y_train,y_test=train_test_split(final,result,test_size=0.2)
clf = LogisticRegression(random_state=0).fit(X_train, y_train)
print('logistic regression score: ',clf.score(X_test,y_test))
```
